```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Merging 2 different datasets to create a new dataset.

```python
credits = (pd.read_csv(r"credits.csv"))
credits
```

```python
names_dict = {}
for id in credits['id'].unique():
    names_list = credits[credits['id'] == id]['name'].tolist()

    names_dict[id] = names_list

names_dict
```

```python
movies = (pd.read_csv(r"titles.csv"))
movies
```

```python
combined_df = pd.DataFrame({'id': list(names_dict.keys()), 'Actors': list(names_dict.values())})

combined_df
```

```python
merged_df = pd.merge(movies, combined_df, on='id', how='inner')
movies = merged_df
movies
```

```python
movies.shape
```

```python
movies.info()
```

Removing features which we do not require for our project.

```python
movies.isnull().sum().sort_values(ascending=False)
```

```python
movies.isnull().sum(axis=1).sort_values(ascending=False)
```

```python
movies.isnull().sum(axis=0).sort_values(ascending=False)/len(movies)*100
```

```python
movies=movies.drop([
  'description',
  'age_certification',
  'seasons'
],axis=1)
```

```python
movies.info()
```

Removing empty datasets which have greater null values than 1%.

```python
round(movies.isnull().sum().sort_values(ascending=False)/len(movies)*100,2)
```

```python
movies=movies[movies['imdb_votes'].notnull()]
movies=movies[movies['imdb_score'].notnull()]
movies=movies[movies['imdb_id'].notnull()]
movies=movies[movies['tmdb_score'].notnull()]
```

```python
round(movies.isnull().sum().sort_values(ascending=False)/len(movies)*100,2)
```

Replacing null values in release year with mean values of the rest of the feature set.

```python
mean_year = movies['release_year'].mean()
movies['release_year'] = movies['release_year'].fillna(mean_year).astype(int)
movies
```

```python
movies.info()
```

removing duplicate rows.

```python
columns_to_check = ['id', 'title', 'type', 'release_year', 'runtime',
       'production_countries', 'imdb_id', 'imdb_score', 'imdb_votes',
       'tmdb_popularity', 'tmdb_score']
duplicates_mask = movies.duplicated(subset=columns_to_check, keep='first')

updated_movies = movies[~duplicates_mask]
updated_movies.reset_index(drop=True)
```

replacing Null values in movie title with "Untitled Movie"

```python
updated_movies['title'].fillna("Untitled Movie", inplace=True)
updated_movies
```

Replacing runtime in minutes with runtime in hours.

```python
updated_movies['runtime_hours'] = (updated_movies['runtime'] / 60).round(2)
updated_movies = updated_movies.drop(columns=['runtime'])
updated_movies
```

removing Shows from type and only keeping movies.

```python
updated_movies = updated_movies[updated_movies['type'] == 'MOVIE']
updated_movies
```

removing outliers  i.e. imdb votes less than 100

```python
updated_movies = updated_movies[updated_movies['imdb_votes'] >= 100.0]
```

Performing one hot encoding on feature "genres"

```python
import ast

unique_genres = set([genre for genres_list in updated_movies['genres'] for genre in ast.literal_eval(genres_list)])

cleaned_genres = [genre.replace(' ', '_').replace(',', '').replace('[', '').replace(']', '') for genre in unique_genres]

one_hot_encoded_genres = updated_movies['genres'].apply(lambda x: pd.Series([1 if genre in x else 0 for genre in unique_genres], index=cleaned_genres))

updated_movies = pd.concat([updated_movies, one_hot_encoded_genres], axis=1)

updated_movies.drop('genres', axis=1, inplace=True)
```

# EDA

```python
# Step 1 (Statistical description of all the columns)
updated_movies.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2 (Univariate Analysis of imdb_score in Histogram)
plt.figure(figsize=(8, 6))
plt.hist(updated_movies['imdb_score'], bins=20, edgecolor='k')
plt.xlabel('IMDB Score')
plt.ylabel('Frequency')
plt.title('Distribution of IMDB Scores')
plt.show()
```

```python
# Step 3 (Correlation Matrix of all the features in the dataset)
correlation_matrix = updated_movies.corr()
plt.figure(figsize=(20, 18))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Step 4 (Scatter Plot IMDB Score VS Release Year)
plt.figure(figsize=(8, 6))
plt.scatter(updated_movies['release_year'], updated_movies['imdb_score'], alpha=0.5)
plt.xlabel('Release Year')
plt.ylabel('IMDB Score')
plt.title('IMDB Score vs. Release Year')
plt.show()
```

```python
# Step 5 (Count Plot of Movies of Comedy Genre)
plt.figure(figsize=(12, 6))
sns.countplot(data=updated_movies, x='comedy')
plt.xticks(rotation=90)
plt.xlabel('Comedy Genre')
plt.ylabel('Count')
plt.title('Count of Movies by Comedy genre')
plt.show()
```

```python
# Step 6 (Bar Plot of mean IMDB score by Release Year)
genre_mean_scores = updated_movies.groupby('release_year')['imdb_score'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_mean_scores.index, y=genre_mean_scores.values)
plt.xticks(rotation=90)
plt.xlabel('Genre')
plt.ylabel('Average IMDB Score')
plt.title('Average IMDB Score by Release Year')
plt.show()
```

```python
# Step 7 (Bar Plot showing specific genre and and the mean of their IMDB votes)
genre_avg_votes = {}
genre_columns = ['action', 'animation', 'comedy', 'crime', 'documentation', 'drama', 'european', 'family', 'fantasy', 'history', 'horror', 'music', 'romance', 'reality', 'sport', 'scifi', 'thriller', 'war', 'western']

for genre in genre_columns:
    genre_avg_votes[genre] = updated_movies[updated_movies[genre] == 1]['imdb_votes'].mean()

plt.figure(figsize=(12, 6))
plt.bar(genre_avg_votes.keys(), genre_avg_votes.values(), color='skyblue')
plt.xlabel('Genre')
plt.ylabel('Average IMDB Votes')
plt.title('Average IMDB Votes by Genre')
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 8 (Pie Chart showing distribution of movies by genres)
genres = ['reality', 'comedy', 'documentation', 'european', 'family', 'action', 'history', 'thriller', 'sport', 'romance', 'animation', 'war', 'crime', 'western', 'drama', 'scifi', 'music', 'horror', 'fantasy']

genre_counts, genre_labels = [], []

for genre in genres:
    genre_count = updated_movies[updated_movies[genre] == 1][genre].count()
    genre_counts.append(genre_count)
    genre_labels.append(genre)

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(genre_counts, labels=genre_labels, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Distribution of Movies in Different Genres')
plt.show()
```

```python
# Step 9 (Univariate Analysis of runtime hours in Histogram)
plt.figure(figsize=(8, 6))
plt.hist(updated_movies['runtime_hours'], bins=20, edgecolor='k')
plt.xlabel('Runtime in Hours')
plt.ylabel('Frequency')
plt.title('Distribution of Movies by Runtime (Hours)')
plt.show()
```

```python
# Step 10 (Scatter Plot comparing Runtime in hours by IMDB Score)
plt.figure(figsize=(10, 6))
plt.scatter(updated_movies['runtime_hours'], updated_movies['imdb_score'], alpha=0.5, color='lightgreen')
plt.xlabel('Runtime (Hours)')
plt.ylabel('IMDB Score')
plt.title('IMDB Score vs. Runtime')
plt.grid(True)
plt.show()
```
